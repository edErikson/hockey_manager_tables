#### Calculate goals and assists for each player and create df

 load csv file to get player names

```python
player_names = pd.read_csv('clean_data/players.csv')
player_names['name'][:3]
```

### Save to list goal scorers and their assistants
#### away games

```python
some=[]
for i in player_names['name']:
    goals = logf[logf['home_team'].str.startswith(i,na=False)]
    name = goals['home_team'][goals.index].str.split('-',1)
    some.append(name)
```

```python
some[0]
```

#### now create dict with key as player and items as assistants
    - then dict to dataframe

```python
away_g_scorers = {}
```

```python
for i in range(len(some)):
    for key,value in some[i]:
        if key not in away_g_scorers:
            away_g_scorers[key] = [value]
            type(away_g_scorers[key]) == list
        elif type(away_g_scorers[key]) == list:
            away_g_scorers[key].append(value)
        else:
            away_g_scorers[key] = [away_g_scorers[key], value]
```

```python
goal_a = pd.DataFrame.from_dict(away_g_scorers,orient='index')
```

```python
goal_a[:2]
```

#### Add pp (powerplay) and sh(shorthanded) goal count at away games, unassisted aswell

```python
goal_a = goal_a.reset_index()
```

new dict for final away games calculations

```python
fin_a = {}
```

```python
for i in range(len(goal_a)):
    name = goal_a['index'][i]
    total_goals = goal_a.iloc[i][goal_a.iloc[i].str.contains(':',na=False)].count()
    pp_goals = goal_a.iloc[i][goal_a.iloc[i].str.contains('PP',na=False)].count()
    sh_goals = goal_a.iloc[i][goal_a.iloc[i].str.contains('SH',na=False)].count()
    un_goals = goal_a.iloc[i][goal_a.iloc[i].str.contains('unassisted',na=False)].count()
    
    goals = [pp_goals,sh_goals,un_goals,total_goals]
    if name not in fin_a:
        fin_a[name] = goals
```

```python
fin_a
```

```python
# away_table dataframe for away scoring
```

```python
away_table = pd.DataFrame.from_dict(fin_a,orient='index',
                                    columns=['Away_pp','Away_sh','A_un','A_goals'])
```

# Player assists

#### use previous data

    - goal_sc   for Home games
    - goal_a    for away games

## Home games

### if player assisted more than once than just add +1 to value

```python
home_assists = {}
```

```python
for i in range(len(goal_sc)):
    for nam in player_names['name']:
        
        counter = goal_sc.iloc[i][goal_sc.iloc[i].str.contains(nam,na=False)].count()
        name = goal_sc['index'][i].strip()
        
        data = [nam,counter]
        
        if name not in home_assists:
            if counter > 0 and name != nam:
                home_assists[name] = [data]
        else:
            if counter > 0 and name != nam:
                home_assists[name].append(data)
```

```python
home_as = pd.DataFrame.from_dict(home_assists,orient='index')
```

```python
home_as[:3]
```

### counting total assists for each player (home games)

```python
home_as = home_as.reset_index()
```

```python
assist_counter = {}
```

```python
for i in range(len(home_as)):
    for goal in range(len(home_as.iloc[0])):
        for name in player_names['name']:
            try:
                if name in home_as.iloc[i][goal][0]:
                    plr_name = home_as['index'][i]
                    
                    counter = home_as.iloc[i][goal][1]
                    if name not in assist_counter:
                        assist_counter[name] = [counter]
                    elif name in assist_counter:
                        assist_counter[name].append(counter)
            except:
                TypeError
```

```python
assisters = pd.DataFrame.from_dict(assist_counter,orient='index')
```

```python
assisters[:3]
```

#### count total home assists

```python
assisters['H_assist'] = assisters.sum(axis=1)
```

```python
assisters = assisters.sort_values(by=['H_assist'],ascending=False)
```

```python
assisters[:5]
```
