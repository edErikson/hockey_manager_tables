```python
import pandas as pd
```

### work with raw csv file - HA_team_player_stats.csv
#### Tasks:
1. add sign like:   
                - 'W' - for attacking players
                - 'D' - for defensive players
                - 'C' - for center players

```python
cols = ['name','age','qua','pot','goa','def','att','sho','pas','spe','str','sco','form','exp','ai','week']
data = pd.read_csv('raw_data/HA_team_player_stats.csv',names=cols)
data[:5]
```

```python
data.info()
```

#### change  'pot' and 'qua' values to int type.
it's not necessary, but maybe in future will need them as integer type!




```python
data[['qua','pot']] = data[['qua','pot']].astype(int)
```

- add new column pos (position)
- insert values to a new column player position - D (defence) W (winger), C (center)
- save to clean_data/players_table.csv

p.s. don't include goalies and columns like - [goa,form,week]


##### Drop [goa,form,week] and goalies from data

```python
goalie = data[data['goa']>0]
data = data.drop(index=goalie.index)
data = data.drop(['goa','form','week'],axis=1)
data[:4]
```

### table with defensive players

```python
def_ = data[data['def'] > data['att']]
def_ = def_.sort_values(by=['ai','age'],ascending=[False,True])
def_ = def_.reset_index(drop=True)
def_.index = def_.index + 1
def_
```

```python
def_.to_csv('clean_data/defence_plr.csv')
```

### wingers

```python
wing = data[(data['att'] > data['def']) & (data['sho'] > data['pas']) & (data['str'] < 20)]
wing = wing.sort_values(by=['ai','age'],ascending=[False,True])
wing = wing.reset_index(drop=True)
wing.index = wing.index + 1
wing[:4]
```

```python
wing.to_csv('clean_data/wingers_plr.csv')
```

### centers

```python
centers = data[(data['att'] > 50) & (data['pas'] > 20) & (data['str'] > 20)]
centers = centers.sort_values(by=['ai','age'],ascending=[False,True])
centers = centers.reset_index(drop=True)
centers.index = centers.index + 1
centers
```

```python
centers.to_csv('clean_data/center_plr.csv')
```

### Concat Winger,Defence,Center tables in one dataframe
##### add column [pos] - with W,D,C initials

```python
wing['pos'] = 'W'
def_['pos'] = 'D'
centers['pos'] = 'C'
```

```python
players_table = pd.concat([wing,def_,centers])
plr_sorted = players_table.sort_values(by=['ai','age'],ascending=[False,True])
```

```python
plr_sorted = plr_sorted.reset_index(drop=True,col_level=1)
```

```python
total_players = plr_sorted[['name','pos','age','ai']]
```

```python
total_players.index = total_players.index + 1
```

```python
total_players.to_csv('clean_data/players.csv')
```
