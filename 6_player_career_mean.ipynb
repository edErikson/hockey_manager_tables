```python
total_list = []
```

```python
def means():
    b = 0
    for i in range(0,27):
        player_one = final[player.index[b]:player.index[b+1]]
        pos = player.index[b]
        
        total = player_one[['Mgames','Mshots','Mgoals','Masists','Mgame_coef','Minj']].agg(['mean'])
        total['name'] = player_one['name'][pos]
        total['seasons'] = player.index[b+1] - player.index[b]
        total_list.append(total)
        b += 1
```

```python
means()
```

```python
len(total_list)
```

```python
pd.DataFrame(total_list[0])
```

```python
calculations = pd.concat([total_list[x] for x in range(27)])
```

```python
sorted_calc = calculations.sort_values(by=['Mgoals','Masists'],ascending=False)
```
